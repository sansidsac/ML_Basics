```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df=pd.read_csv("data.csv")
df
```

```python
import math
median_bedrooms=math.floor(df.bedrooms.median())
median_bedrooms
```

```python
df.bedrooms=df.bedrooms.fillna(median_bedrooms)
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df['price'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
def variables(ar,bd,ag):
    return pd.DataFrame({'area':[ar],'bedrooms':[bd],'age':[ag]})
```

```python
reg.predict(variables(3000,3,40))
```

```python
reg.predict(variables(2500,5,4))
```
