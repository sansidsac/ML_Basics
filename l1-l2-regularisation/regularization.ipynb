```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataset=pd.read_csv('Melbourne_housing_FULL_acedit.csv')
dataset.head()
```

```python
dataset.nunique()
```

```python
dataset.shape
```

```python
cols_to_use=['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance',
             'CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
dataset=dataset[cols_to_use]
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_zero=['Propertycount','Distance','Bedroom2','Bathroom','Car']
dataset.loc[:,cols_to_fill_zero]=dataset[cols_to_fill_zero].fillna(0)
dataset.isna().sum()
```

```python
dataset['Landsize']=dataset['Landsize'].fillna(dataset['Landsize'].mean())
dataset['BuildingArea']=dataset['BuildingArea'].fillna(dataset['BuildingArea'].mean())
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
dataset.isna().sum()
```

```python
dataset=pd.get_dummies(dataset, drop_first=True)
dataset.head()
```

```python
X=dataset.drop(['Price'],axis=1)
y=dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression().fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.score(X_train,y_train)
```

```python
from sklearn.linear_model import Lasso

lasso_reg=Lasso(alpha=50, max_iter=1000, tol=0.1)

lasso_reg.fit(X_train,y_train)
```

```python
lasso_reg.score(X_test,y_test)
```

```python
lasso_reg.score(X_train,y_train)
```

```python
from sklearn.linear_model import Ridge

ridge_reg=Ridge(alpha=50, max_iter=1000, tol=0.1)

ridge_reg.fit(X_train,y_train)
```

```python
ridge_reg.score(X_test,y_test)
```

```python
ridge_reg.score(X_train,y_train)
```
