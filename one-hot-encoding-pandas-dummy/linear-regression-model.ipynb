```python
import pandas as pd
df=pd.read_csv('data.csv')
df
```

```python
dummies=pd.get_dummies(df['town'])
dummies=dummies.astype('int')
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
final=merged.drop(['town','west windsor'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
X=final.drop('price',axis='columns')
X
```

```python
Y=final['price']
Y
```

```python
model.fit(X,Y)
```

```python
def predict(a,m,r):
    return pd.DataFrame({'area':[a],'monroe township':[m],'robinsville':[r]})
```

```python
model.predict(predict(2800,0,1))
```

```python
model.predict(predict(3300,0,0))
```

```python
model.score(X,Y)
```
